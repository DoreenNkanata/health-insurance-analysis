```python
# Step 1: Map insurance status to binary
data['has_insurance'] = data['have_you_ever_had_health_insurance?'].map({'Yes': 1, 'No': 0})

# Step 2: Group by employment_status and insurance
grouped = data.groupby(['employment_status', 'has_insurance']).size().unstack(fill_value=0)

# Step 3: Convert to percentage
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Step 4: Plot side-by-side bar chart
ax = percentages.plot(kind='bar', color=['blue', 'mediumseagreen'], width=0.7, figsize=(10, 6)) # Set the desired figure size (width, height)
plt.title('Insurance Coverage by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Insurance Access in Percentage')
plt.legend(['No Insurance (0)', 'Has Insurance (1)'], title='Insurance Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###  **Interpretation: Insurance Coverage by Employment Status**

This bar chart displays the percentage of people with and without health insurance, grouped by **employment status**: Employed, Self-employed, and Unemployed. The green bars represent individuals **with insurance**, while blue bars indicate those **without insurance**.


###  **Key Insights:**

1. **Employed individuals have the highest insurance coverage**:

   * 79% of employed people are insured.
   * Only 21% of employed individuals lack insurance.
   * This suggests that formal employment is strongly linked to better access to health insurance,possibly due to employer-provided benefits or stable income.

2. **Self-employed individuals show a more balanced but lower coverage rate**:

   * 62% have insurance, while 38% don’t.
   * Self-employment might offer flexibility but comes with reduced access to employer-based insurance schemes.

3. **Unemployed individuals are the most uninsured**:

   * 66% of unemployed respondents lack health insurance.
   * Only 34% have coverage, highlighting a significant gap for this vulnerable group.



###  **Recommendations:**

1. **Expand public insurance programs for the unemployed**:

   * Introduce subsidized or government-sponsored health insurance for people without jobs.
   * Work with NGOs and community health organizations to fill the access gap.

2. **Create insurance incentives for the self-employed**:

   * Offer tax breaks or group plans for freelancers, small business owners, and gig workers to increase their coverage.

3. **Strengthen employment-linked health benefits**:

```python
#Map cancer screening responses to binary
data['cancer_screening'] = data['have_you_ever_had_a_cancer_screening_(e.g._mammogram,_colonoscopy,_etc.)?'].map({'Yes': 1, 'No': 0})
```

```python
#Drop rows with missing gender, age, or screening response
filtered = data.dropna(subset=['gender', 'age', 'cancer_screening'])
```

```python
#Group by age and gender to calculate screening percentages
grouped = (
    filtered.groupby(['age', 'gender'])['cancer_screening']
    .mean()
    .unstack()
    .fillna(0) * 100
)
```

```python
#Plot side-by-side bar chart
grouped.plot(kind='bar', figsize=(10, 6))
plt.title('Cancer Screening Rates by Age and Gender')
plt.ylabel('Percentage Screened')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

###  **Interpretation of the Graph: Cancer Screening Rates by Age and Gender**

The bar chart presents the **percentage of individuals screened for cancer** across **different age groups**, separated by **gender (Female vs. Male).**


###  **Key Insights:**

1. **Cancer screening increases with age** for both genders:

   * Younger adults (18–30) show the **lowest screening rates**, especially males (≈10%) and females (≈14%).
   * Screening rates peak in the **60+ age group**, with **females at 60%** and **males around 35%**.

2. **Females consistently screen more than males** across all age groups:

   * The gender gap in screening becomes **more significant with age**, especially in the 60+ group (25%+ difference).
   * This may be due to more routine cancer screening programs targeting women (e.g., breast and cervical cancer), while men may lack equivalent awareness or access.

3. The **screening gap narrows slightly** in the 51–60 group, suggesting targeted interventions may have some effect, but disparities persist.


###  **Recommendations:**

1. **Target Younger Age Groups with Awareness Campaigns:**

   * Encourage early screening through schools, universities, and young adult-focused platforms.
   * Emphasize early detection and its life-saving potential.

2. **Increase Male Screening Participation:**

   * Launch **male-focused cancer awareness** programs (e.g., for prostate and colorectal cancer).
   * Collaborate with male community leaders, workplaces, and health influencers to break stigma and misinformation.

3. **Sustain and Expand Female Screening Programs:**

   * Continue supporting initiatives like HPV vaccination, mobile mammography units, and community health drives.
   * Ensure older women (60+) remain a focus as risk increases with age.

4. **Policy Integration:**

   * Integrate cancer screening into general health check-ups at public facilities, especially for people 40+.
   * Offer **subsidized or free screening** to low-income groups across both genders.


```python
#1. Convert hospital visit column to numeric
data['months_since_last_visit'] = pd.to_numeric(
    data['when_was_the_last_time_you_visited_a_hospital_for_medical_treatment?_(in_months)'],
    errors='coerce'
)
```

```python
#2. Drop missing values for relevant columns
visit_df = data.dropna(subset=['gender', 'age', 'months_since_last_visit'])
```

```python
#3. Group by age and gender, compute average time since last visit
grouped_visits = (
    visit_df.groupby(['age', 'gender'])['months_since_last_visit']
    .mean()
    .unstack()
    .fillna(0)
)
```

```python
#4. Plot the results
grouped_visits.plot(kind='bar', figsize=(10, 6), colormap='coolwarm')
plt.title('Average Months Since Last Hospital Visit by Age and Gender')
plt.ylabel('Average Months Since Last Visit')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

###  Interpretation of the Graph:

**"Average Months Since Last Hospital Visit by Age and Gender"**

This bar chart visualizes how frequently different age groups of males and females visit the hospital for medical treatment, based on **average months since the last visit** (i.e., higher values = less frequent visits).

---

###  Key Insights:

1. **18–30 Age Group:**

   * **Males** ( 9 months) go the longest without hospital visits across all groups.
   * **Females** ( 7 months) visit slightly more frequently than males.

2. **31–60 Age Groups:**

   * Both genders show a **gradual increase in hospital visits** (fewer months between visits) as age increases.
   * **Females consistently visit hospitals more frequently than males** in each group (by \~1 month).

3. **60+ Age Group:**

   * This group has the **shortest average time since last visit** for both genders.
   * Indicates **increased medical attention** or monitoring in older age.
   * Again, **females** visit more frequently than **males**.

---

###  Interpretation Summary:

* **Young adult males (18–30)** are the least likely to seek medical treatment regularly.
* **Females across all age groups** engage more frequently with healthcare services, potentially due to:

  * Reproductive health needs
  * Preventive care behavior
  * Higher health awareness
* **Older adults (60+)** have more regular hospital visits, likely due to age-related conditions.

   * Promote routine check-ups and preventive screenings.

2. **Workplace & School Health Drives:**
